```python
from pandas import read_csv
df = read_csv(r"J:\USGS_Field_Survey\FAM_Project_Data-Analysis_2012_2016\FAM_Survey2017\test_merge_csv_in_python\reclassified_2017.csv") #Reads and opens csv file.
df #Call dataframe and display contents of csv.
```

```python
df.columns = ['SIMS_ID', 'S20170922', 'S20170524'] #Changes column headers or even assigns them.
```

```python
df['S20170524'].replace([4],[10],inplace=True) #Replaces values of Emergent code 4 for winter with Idle code 10.
df['S20170524'].replace([0],['NAN'],inplace=True) #Replaces values of 0 for winter with NaN.
df['S20170922'].replace([0],['NAN'],inplace=True) #Replaces values of 0 for summer with NaN.
df
```

```python
df['YTD20170922'] = '10' #Following df.loc[(df[columnheader])] similarly use theory of conditional arguments to populate the'YTDxxx" column. reclassified code 2 == Cropped and code 10 == Idle.
df
df.loc[(df['S20170922'] > 10) & (df['S20170524'] > 10), 'YTD20170922'] = '10'
df
df.loc[df['S20170922'] < 10,'YTD20170922'] = '2'
df
df.loc[df['S20170524'] < 10,'YTD20170922'] = '2'
df
df.loc[df['S20170922'] == "NAN",'YTD20170922'] = "NAN"
df
```

```python
df.to_csv(r"J:\USGS_Field_Survey\FAM_Project_Data-Analysis_2012_2016\FAM_Survey2017\test_merge_csv_in_python\test_conditionals.csv", index = False)
#Writes edited csv to new csv file.
```
